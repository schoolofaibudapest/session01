## Bitcoin:
### Who created it?
> Satoshi Nakamoto proposed bitcoin in 2008, as an electronic payment system based on mathematical proof. The idea was to produce a means of exchange, independent of any central authority, that could be transferred electronically in a secure, verifiable and immutable way.
Bitcoin can be used to pay for things electronically, if both parties are willing. In that sense, it's like conventional dollars, euros, or yen, which are also traded digitally.

### In what ways is it different from traditional currencies?
> Bitcoin can be used to pay for things electronically, if both parties are willing. In that sense, it's like conventional dollars, euros, or yen, which are also traded digitally.

### But it differs from fiat digital currencies in several important ways:

#### Decentralization:
> Bitcoin's most important characteristic is that it is decentralized. No single institution controls the bitcoin network. It is maintained by a group of volunteer coders, and run by an open network of dedicated computers spread around the world. This attracts individuals and groups that are uncomfortable with the control that banks or government institutions have over their money.
Bitcoin solves the "double spending problem" of electronic currencies (in which digital assets can easily be copied and re-used) through an ingenious combination of cryptography and economic incentives. In electronic fiat currencies, this function is fulfilled by banks, which gives them control over the traditional system. With bitcoin, the integrity of the transactions is maintained by a distributed and open network, owned by no-one.

#### Limited supply:
> Fiat currencies (dollars, euros, yen, etc.) have an unlimited supply – central banks can issue as many as they want, and can attempt to manipulate a currency's value relative to others. Holders of the currency (and especially citizens with little alternative) bear the cost.
With bitcoin, on the other hand, the supply is tightly controlled by the underlying algorithm. A small number of new bitcoins trickle out every hour, and will continue to do so at a diminishing rate until a maximum of 21 million has been reached. This makes bitcoin more attractive as an asset – in theory, if demand grows and the supply remains the same, the value will increase.

#### Pseudonymity:
> While senders of traditional electronic payments are usually identified (for verification purposes, and to comply with anti-money laundering and other legislation), users of bitcoin in theory operate in semi-anonymity. Since there is no central "validator," users do not need to identify themselves when sending bitcoin to another user. When a transaction request is submitted, the protocol checks all previous transactions to confirm that the sender has the necessary bitcoin as well as the authority to send them. The system does not need to know his or her identity.
In practice, each user is identified by the address of his or her wallet. Transactions can, with some effort, be tracked this way. Also, law enforcement has developed methods to identify users if necessary.
Furthermore, most exchanges are required by law to perform identity checks on their customers before they are allowed to buy or sell bitcoin, facilitating another way that bitcoin usage can be tracked. Since the network is transparent, the progress of a particular transaction is visible to all.
This makes bitcoin not an ideal currency for criminals, terrorists or money-launderers.


#### Immutability:
> Bitcoin transactions cannot be reversed, unlike electronic fiat transactions.
This is because there is no central "adjudicator" that can say "ok, return the money." If a transaction is recorded on the network, and if more than an hour has passed, it is impossible to modify.
While this may disquiet some, it does mean that any transaction on the bitcoin network cannot be tampered with.


#### Divisibility:
>The smallest unit of a bitcoin is called a satoshi. It is one hundred millionth of a bitcoin (0.00000001) – at today's prices, about one hundredth of a cent. This could conceivably enable microtransactions that traditional electronic money cannot.

```python
# Import the libraries we will use.

import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_validate
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression
```

```python
# Source the dataset for the model.
# Note that: This could be different on your environment.

# My dataset is in this DIR
dataset_location = '/Users/horvathpeter/SchoolOfAI/Session01_Regression/dataset/bitcoin-historical-data'
# We use only bitstamp database
dataset_name = 'bitstampUSD_1-min_data_2012-01-01_to_2018-06-27.csv'
# Merge the dataset DIR, and the dataset name. /Make it future proof./
df = pd.read_csv("{}/{}".format(dataset_location, dataset_name))
```

```python
# Let's see what we got here.
# Print the header and some random rows from the dataset.

random_subset = df.sample(n=5)
print(random_subset.head())
```

```python
# Let's visualize some of our data's columns
# To visualize in jupyter notebook we the matplotlib notebook also
%matplotlib notebook 
fig, axes = plt.subplots(nrows=2, ncols=2)
df['Open'].plot(ax=axes[0,0]); axes[0,0].set_title('Open')
df['Close'].plot(ax=axes[0,1]); axes[0,1].set_title('Close')
df['Volume_(BTC)'].plot(ax=axes[1,0]); axes[1,0].set_title('Volume_(BTC)')
df['Volume_(Currency)'].plot(ax=axes[1,1]); axes[1,1].set_title('Volume_(Currency)')
plt.subplots_adjust(wspace=0.5, hspace=0.5)
plt.show()
```

```python
# For later usage, it's a good idea to define what column we want to predict.
# Here we want to predict the 'Close' stock price.
# FORECAST_PCT define the percentage we want to predict from the database.

FORECAST = 'Close'
FORECAST_PCT = 0.009
```

```python
# We do not want 'Nan's in our dataset. That will lead errors at the learning stage.
# Depends on your dataset there are multiple choices.
# We will replace 'Nan' with the last know value here if any.
# In ML we have to handle this problem, but we also want to keep as much data as possible.

# Fill Nans with last know value in the columns

# df[df.columns.values] = df[df.columns.values].ffill()
# Drop any Nans from the dataframe if any.
# df.dropna(inplace=True)
```

```python
# For large dataset we can also use only the 'n'th datas from the dataframe. We use all of them now.
n = 1
df = df[df.index % n == 0]
```

```python
# Let's define some usefull columns as well.
# OCPCT is the Open/Close percentage difference, that shows us how stable the price is.
# HLPCT is the High/Low percentage difference, that also shows use the movement of the price.

df['OCPCT'] = ((df['Close'] - df['Open']) / df['Open']) * 100
df['HLPCT'] = ((df['High'] - df['Low']) / df['Close']) * 100
```

```python
# Redefine the columns of our the dataframe. We want here just the usable datas.

df = df[['OCPCT', 'HLPCT', 'Volume_(BTC)', 'Close',]]
```

```python
# Let's see what we got after the reconstruction.
# Print the header and some random rows from the dataset.

random_subset = df.sample(n=5)
print(len(df))
print(df.head())
print(random_subset.head())
```

```python
# # Let's visualize the columns we will use.

%matplotlib notebook
fig, axes = plt.subplots(nrows=2, ncols=2)
df['Close'].plot(ax=axes[0,0]); axes[0,0].set_title('Close')
df['Volume_(BTC)'].plot(ax=axes[0,1]); axes[0,1].set_title('Volume_(BTC)')
df['OCPCT'].plot(ax=axes[1,0]); axes[1,0].set_title('OCPCT')
df['HLPCT'].plot(ax=axes[1,1]); axes[1,1].set_title('HLPCT')
plt.subplots_adjust(wspace=0.5, hspace=0.5)
plt.show()
```

```python
# Forecast out a certain number of datapoints
# Calculate the mean of the percentage
# For now we want to forcast out the last FORECAST_PCT from the dataset

forecast_out = int(math.ceil(FORECAST_PCT*len(df)))
print(forecast_out)
```

```python
# Define the 'label' for dataframe.
# 'label' shall be the price, in the future, where the future is FORECAST_PCT% of the entire length of the dataset out
df_close = df[['Close']].copy()
print(len(df_close))
df['label'] = df[FORECAST].shift(-forecast_out)
```

```python
# Defined X (features), as our entire dataframe except for the label column, converted to a numpy array.
# We do this using the .drop method that can be applied to dataframes, which returns a new dataframe.
X = np.array(df.drop(['label'], 1))

# Preprocessing makes our features range between -1 and 1
X = preprocessing.scale(X)

# Define X_lately from our dataset. This will be the datas we want to predict in the future.
X_lately = X[-forecast_out:]

# The other datas we want to use for train, and testing.
X = X[:-forecast_out]

# Drop any Nans from the dataframe if any.
df.dropna(inplace=True)

# Next, we define our y variable, which is our label, as simply the label column of the dataframe, 
# converted to a numpy array.
y = np.array(df['label'])

# We split our remaining dateframe to two pieces.
# One is our training set
# The remaining is our test set for classifier
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, shuffle=False)

# Define our classifier. For now we use LinearRegression, but in sklearn you can find others too.
clf = LinearRegression(n_jobs=-1)

# Train our model with sklearn. We do this with .fit method
clf.fit(X_train, y_train)

# Check our model's accuracy on the setset.
accuracy = clf.score(X_test, y_test)
print("Model accuracy {}".format(accuracy))
```

```python
# Also we want to predict some forecasts.
# This will show you the model accuracy on a not known input (X_lately in this case)
forecast_set = clf.predict(X_lately)

print(forecast_set)
#print(df_close.loc[len(df_close)-forecast_out:len(df_close)])

```
